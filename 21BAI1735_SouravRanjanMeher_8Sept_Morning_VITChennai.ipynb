```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
car = sb.load_dataset('car_crashes')
```

```python
car.head()
```

```python
car
```

```python
#lineplot
sb.lineplot(data=car, x="total", y="alcohol")
plt.title("Car Crashes vs. Alcohol Consumption")
plt.xlabel("Total Car Crashes")
plt.ylabel("Alcohol Consumption")
plt.show()
```

```python
#Subolots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sb.lineplot(data=car, x="total", y="alcohol", ax=axes[0])
axes[0].set_title("Car Crashes vs. Alcohol Consumption")
axes[0].set_xlabel("Total Car Crashes")
axes[0].set_ylabel("Alcohol Consumption")
sb.scatterplot(data=car, x="speeding", y="alcohol", ax=axes[1])
axes[1].set_title("Speeding vs. Alcohol Consumption")
axes[1].set_xlabel("Speeding")
axes[1].set_ylabel("Alcohol Consumption")
plt.tight_layout()
plt.show()
```

```python
#ScatterPlot
sb.scatterplot(data=car, x="speeding", y="total")
plt.title("Scatter Plot of Car Crashes vs. Speeding")
plt.xlabel("Speeding")
plt.ylabel("Total Car Crashes")
plt.show()
```

```python
#BarPlot
plt.figure(figsize=(10, 8))
sb.barplot(data=car, x="abbrev", y="total", palette="Blues", ci=None, saturation=1.0, edgecolor=".2")
plt.title("Total Car Crashes by State ")
plt.xlabel("State")
plt.ylabel("Total Car Crashes")
plt.xticks(rotation=45)
plt.show()
```

```python
#Histogram
sb.histplot(data=car, x="total", bins=20, kde=True, color="skyblue")
plt.title("Histogram of Total Car Crashes")
plt.xlabel("Total Car Crashes")
plt.ylabel("Frequency")
plt.show()
```

```python
#PieChart

alcohol_accidents = (car['alcohol'] > 0).sum()
speeding_accidents = (car['speeding'] > 0).sum()
labels = ['Alcohol', 'Speeding']
sizes = [alcohol_accidents, speeding_accidents]
colors = ['lightcoral', 'lightskyblue']
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Car Crashes by Causes")
plt.show()
```

```python
#jointplot
sb.set(style="white")
#Relationsship between alcohol and speeding
sb.jointplot(data=car, x="speeding", y="alcohol", kind="scatter", color="skyblue")
```

```python
#BoxPlot
plt.figure(figsize=(8, 6))  # Set figure size
sb.boxplot(data=car, y="ins_premium", color="lightblue")
plt.title("Distribution of Insurance Premiums")
plt.ylabel("Insurance Premium")
plt.show()
```

```python
#HEATMAP
correlation_matrix = car.corr()
plt.figure(figsize=(10, 8))
sb.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
